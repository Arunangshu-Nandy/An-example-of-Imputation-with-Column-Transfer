```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
data=pd.DataFrame({'A':(1,np.nan,3),'B':(5,7,np.nan),'C':(np.nan,9,10)})
data
```

```python
impute=SimpleImputer(strategy='mean')
x=impute.fit(data)
x=impute.transform(data)
x
```

```python
# Create a radnom datset of 10 rows and 4 columns
df = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))

# Randomly set some values as null
df = df.mask(np.random.random((10, 4)) < .15)

# Duplicate two cells with same values
df['B'][8] = df['B'][9]
df
```

```python
mean_imputer = SimpleImputer(strategy='mean')

result_mean_imputer=mean_imputer.fit_transform(df)

pd.DataFrame(result_mean_imputer, columns=list('ABCD'))
```

```python
from sklearn.compose import ColumnTransformer

column_trans = ColumnTransformer(
[('A', SimpleImputer(strategy='mean'), [0]),
 ('B', SimpleImputer(strategy='constant', fill_value=29), [1])],
remainder='passthrough')

X=pd.DataFrame(column_trans.fit_transform(df),columns=list('ABCD'))
X
```

```python
column_trans = ColumnTransformer(
[('A', SimpleImputer(strategy='mean'), [0]),
 ('B', SimpleImputer(strategy='constant', fill_value=50), [1]),
('C',SimpleImputer(strategy='most_frequent'),[2])],
remainder='passthrough')

Y=pd.DataFrame(column_trans.fit_transform(df),columns=list('ABCD'))
Y
```
